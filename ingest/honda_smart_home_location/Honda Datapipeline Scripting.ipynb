```python
df = pd.read_csv('/Users/slz/git_repo/tsdat_template_new/ingest/honda_smart_home_location/input_data/Honda_AllData0421_analyzed.csv')
```

```python
template = """

{0}: 
  input:
    name: "{0}" 
  dims:
    [time] 
  type: {1} 
  attrs:
    long_name: {0} 
    standard_name:
      "" 
    comment: "" 
    units:
      1
"""
for i in range(0, len(df.columns)):
    data_type = df.dtypes[i]
    if data_type=="float64":
        data_type = "float"
    if df.columns[i]!="Timestamp":
        print(template.format(df.columns[i], data_type).strip()+"\n")
```
